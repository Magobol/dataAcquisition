```python
import requests as r
from bs4 import BeautifulSoup
import re
import re
regex = re.compile(r'http.*|@.*')



github = 'https://gist.github.com/paulmillr/2657075'
def liste_refs(link):
      
    data = r.get(link).content
    soup = BeautifulSoup(data, 'html.parser')
    l=[]
    
#     for s in soup.find_all(class_='mw-parser-output')[0].select('a'):
#         a = s.get('href')
#         l.append(str(a))
    
#     regex = re.compile('\/wiki\w*')
#     tmp  = [s.getText().split(" ")[0] for s in soup.find_all('td')] 
    tmp = soup.find_all('td')
    
    u = []
    for s in tmp:
        try:u.append(s.find('a').getText())
        except:continue
        try:u.remove('')
        except: continue
            
    filtered = [i for i in u if not regex.match(i)]
    
    return filtered#list(filter(regex.search,l))

users = liste_refs(github)[:-3]
users
```
