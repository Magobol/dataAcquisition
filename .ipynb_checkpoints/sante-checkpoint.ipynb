```python
import pandas as pd
```

```python
df_spe = pd.read_excel("data/honoraires_departement.xls",sheet_name = "Spécialistes",header=None, names=["Proffession","Departement","nb_APE","hono_sans_depassement","depassements","frais_deplacement","tot_hono","hono_sans_depaseement/APE","depassements/APE","frais_de_deplacement_APE","honoraires/APE"])
df_gen = pd.read_excel("data/honoraires_departement.xls",sheet_name = "Dentistes et ODF",header=None, names=["Proffession","Departement","nb_APE","hono_sans_depassement","depassements","frais_deplacement","tot_hono","hono_sans_depaseement/APE","depassements/APE","frais_de_deplacement_APE","honoraires/APE"])
df_dent = pd.read_excel("data/honoraires_departement.xls",sheet_name = "Sages-femmes",header=None, names=["Proffession","Departement","nb_APE","hono_sans_depassement","depassements","frais_deplacement","tot_hono","hono_sans_depaseement/APE","depassements/APE","frais_de_deplacement_APE","honoraires/APE"])
df_sf = pd.read_excel("data/honoraires_departement.xls",sheet_name = "Généralistes et MEP",header=None, names=["Proffession","Departement","nb_APE","hono_sans_depassement","depassements","frais_deplacement","tot_hono","hono_sans_depaseement/APE","depassements/APE","frais_de_deplacement_APE","honoraires/APE"])
df_auxMed = pd.read_excel("data/honoraires_departement.xls",sheet_name = "Auxiliaires médicaux",header=None, names=["Proffession","Departement","nb_APE","hono_sans_depassement","depassements","frais_deplacement","tot_hono","hono_sans_depaseement/APE","depassements/APE","frais_de_deplacement_APE","honoraires/APE"])
```

```python
df = df_spe.append(df_gen).append(df_dent).append(df_sf).append(df_auxMed)
```

```python
# df.reset_index(drop=True).drop()
df = df.loc[~(df["depassements"]=="DEPASSEMENTS (Euros)")]
```

```python
df = df.loc[~(df["depassements"]=="nc")].infer_objects()
df = df.loc[~((df["Proffession"].str.contains("TOTAL"))|(df["Departement"].str.contains("TOTAL")))]
```

```python
df[["Proffession","depassements/APE"]].sort_values("depassements/APE",ascending=False)
```
